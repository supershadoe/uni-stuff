```python
def get_graph_input_with_cost() -> dict[str, list[tuple[int, str]]]:
  nodes = input("Enter values for the nodes: ").strip().split()
  graph = { node: [] for node in nodes }
  print("Enter cost for each edge (use - if not present)")
  for node in nodes:
    costs = input(f"From {node}: ").strip().split()
    graph[node] = list(
        map(
            lambda x: (int(x[0]), x[1]),
            filter(
                lambda x: x[0] != "-",
                map(lambda x: (x[0], x[1]), zip(costs, nodes))
            )
        )
    )
  return graph
```

```python
def get_hsld(graph: dict[str, list[tuple[int, str]]]):
  hsld = {}
  for node in graph:
    hsld[node] = int(input(f"Enter SLD from {node}: "))
  return hsld
```

```python
def reconstruct_path(end: str, parent: dict[str, str]) -> list[str]:
  path = []
  curr = end
  while curr is not None:
    path.append(curr)
    curr = parent[curr]
  return path[::-1]

def a_star(
    graph: dict[str, list[tuple[int, str]]],
    h: dict[str, int],
    start: str,
    end: str,
) -> tuple[list[str], int] | None:
  q = [start]
  parent: dict[str, str] = {}
  parent[start] = None
  g = { node: float('inf') for node in graph }
  f = { node: float('inf') for node in graph }
  g[start] = 0
  f[start] = h[start]

  while q:
    curr = None
    least_f = float('inf')
    for i in q:
      if f[i] < least_f:
        curr, least_f = i, f[i]
    q.remove(curr)
    if curr == end:
      return reconstruct_path(curr, parent), int(g[curr])
    for cost, neighbor in graph[curr]:
      new_g = g[curr] + cost
      if new_g < g[neighbor]:
        parent[neighbor] = curr
        g[neighbor] = new_g
        f[neighbor] = new_g + h[neighbor]
        if neighbor not in q:
          q.append(neighbor)

  return None
```

```python
graph = get_graph_input_with_cost()
hsld = get_hsld(graph)
```

```python
start = input("Enter start node: ")
end = input("Enter end node: ")
```

```python
res = a_star(graph, hsld, start, end)

if res is None:
  print("No path was found")
else:
  path, cost = res
  print(f"Least cost path is {' -> '.join(path)} with a cost of {cost}")
```
